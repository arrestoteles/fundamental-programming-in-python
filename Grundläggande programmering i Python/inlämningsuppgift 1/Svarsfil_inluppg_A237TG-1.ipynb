```python
import csv
kameradata = []    
platsdata = []

# Öppna CSV fil och returnera alla värden i en lista
def read_file(file_name):
    with open (file_name, encoding = 'UTF-8') as file:
        lista = []
        csv_reader = csv.reader(file, delimiter = ';')
        
        for row in csv_reader:
            lista.append(row)
        
        return lista
    
# Anropar funktionen read_file(file_name)
kameradata = read_file('kameraData.csv')
platsdata = read_file('platsData.csv')

# Skriv ut de tre första raderna i var och en av de två listorna
print("\n")
print("Kameradata-filen: ")
print(kameradata[0:3])
        
print("\n")
        
print("Platsdata-filen: ")
print(platsdata[0:3])
print("\n")
```

```python
import matplotlib.pyplot as plt

'''
Beräkna totalt antal fordon 
som passerat vid var och en av 
de olika gällande hastigheterna
'''
def antal_bilar(kamera_data):
    
    # Skapa variabler för summering av mätningarna per hastighet
    hastighet_40 = 0
    hastighet_50 = 0
    hastighet_60 = 0
    hastighet_70 = 0
    hastighet_80 = 0
    hastighet_90 = 0
    
    # Skapa variabel för totalt antal passerade bilar
    summa_bilar = 0

    # Villkor för summering av gällande hastiget samt totalt antal passerade bilar
    for row in kamera_data:
        try:
            if int(row[1]) == 40:
                hastighet_40 += 1
                summa_bilar += 1
            elif int(row[1]) == 50:
                hastighet_50 += 1
                summa_bilar += 1
            elif int(row[1]) == 60:
                hastighet_60 += 1
                summa_bilar += 1
            elif int(row[1]) == 70:
                hastighet_70 += 1
                summa_bilar += 1
            elif int(row[1]) == 80:
                hastighet_80 += 1
                summa_bilar += 1
            elif int(row[1]) == 90:
                hastighet_90 += 1
                summa_bilar += 1
        except:
                pass
    
    # Skriv ut output enligt uppgiftsbeskrivningen
    print('Det finns ' + str(hastighet_40) + ' mätningar där gällande hastighet är 40 km/h')
    print('Det finns ' + str(hastighet_50) + ' mätningar där gällande hastighet är 50 km/h')
    print('Det finns ' + str(hastighet_60) + ' mätningar där gällande hastighet är 60 km/h')
    print('Det finns ' + str(hastighet_70) + ' mätningar där gällande hastighet är 70 km/h')
    print('Det finns ' + str(hastighet_80) + ' mätningar där gällande hastighet är 80 km/h')
    print('Det finns ' + str(hastighet_90) + ' mätningar där gällande hastighet är 90 km/h')
    print('------------------------------------------------------------')
    print('Totalt passerade ' + str(summa_bilar) + ' bilar')
    
    # X och Y koordinater som ska plottas
    x_koord = [40,50,60,70,80,90]
    data_to_plot = [hastighet_40, hastighet_50, hastighet_60, hastighet_70, hastighet_80, hastighet_90]
    
    # Plotta tabellen med hjälp av matplotlib
    plt.bar(x_koord, data_to_plot)
    plt.title('Antal fordon för varje gällande hastighet')
    plt.xlabel('Gällande hastighet')
    plt.ylabel('Antal passerande fordon')
    plt.grid()
    plt.show()
    
# Anropar funktionen antal_bilar(kamera_data)
antal_bilar(kameradata)
```

```python
'''
Räkna ihop antal kameror i varje kommun, 
sorterar resultatet i alfabetisk ordning efter kommun 
och presenterar resultatet som en tabell
'''
def antal_kameror(plats_data):
    # Deklarera dictionary där key = kommun och value = antal kameror
    dict = {}
    
    # Deklarera variabel för summering av samtliga kameror för alla kommuner
    summa_kameror = 0
    
    '''
    Om kommunnamn redan finns i dict, addera antal kameror +1.
    Om kommunnamn ej finns i dict, addera key = <kommun> och addera antal kameror +1.
    Hoppa över första elementet där rubrik är lika med "Kommun".
    '''
    for rad in plats_data:
        if rad[3] == 'Kommun':
            continue
        if rad[3] in dict:
            kommun = rad[3]
            dict[kommun] += 1
            summa_kameror += 1
        else:
            dict[rad[3]] = 1
            summa_kameror += 1
            
    # Sortera dict i alfabetisk ordning
    sorterad_dict = sorted(dict.items())    
    
    # Skriv ut output enligt uppgiftsbeskrivningen
    print(f'Kommun  \t  Antal kameror')
    print(f'-----------------------------------')

    for key, value in sorterad_dict:
        print(f'{key:12} {value:>12}')

    print(f'-----------------------------------')
    print(f'Det finns totalt ' + str(summa_kameror) + ' kameror.')
    
# Anropar funktionen antal_kameror(plats_data)
antal_kameror(platsdata)
```

```python
### deluppgift 4a ###
'''
Ta fram information om sådana fall 
där hastigheten var ett av användaren vald procentsats 
större än "Gällande Hastighet".
'''
def hastighetplus(kamera_data):
    
    # Fråga användaren om procentgräns input
    procentover = input("Söker ut alla poster där den procentuella hastighetsöverträdelsen\n"
          "överskrider gällande hastighet med mer än vald %-sats.\n"
          "Med hur många % skall hastigheten överskrida hastighetsbegränsningen för\n"
          "att visas?: ")
    
    # Skapa ny tom lista
    array = kamera_data.copy()
    plushastighet = []
    
    # Beräkna andelen mätningar över procentgräns och addera till listan plushastighet
    for rad in array:
        if rad[2] == "Hastighet":
            continue
        elif (((int(rad[2]) - int(rad[1])) / int(rad[1])) * 100) > float(procentover):
            rad_kopia = rad.copy()
            plushastighet.append(rad_kopia)
                        
    # Swappa kolumner för varje element i ordning enligt krav
    for i in plushastighet:
        i.reverse()
        i[1], i[2], i[3], i[4] = i[4], i[3], i[2], i[1]
    
    # Sortera listan enligt klockslag
    plushastighet.sort()
    
    # Skriv ut text enligt krav i uppgiftsbeskrivningen
    print("\n")
    print("Det var " + str(len(plushastighet)) + " överträdelser som var mer än " + procentover + "% över gällande hastighet.")
    print("\n")
    print(f'Tid  \t\t  MätplatsID  \t  Gällande Hastighet  \t  Hastighet  \t  Datum')
    print(f'-------------------------------------------------------------------------------------')
    
    # Skriv ut varje kolumn för varje element från listan
    for rad in plushastighet:
        print(f'{rad[0]}   \t   {rad[1]} \t\t {rad[2]}   \t\t     {rad[3]} \t {rad[4]}')
    
    return plushastighet, procentover


### deluppgift 4b ###
import matplotlib.pyplot as plt

'''
Rita ett stapeldiagram utifrån en returnerad 
överträdelselista plushastighet och 
överträdelseprocensats procentover 
'''
def hastighetplusdiagram(plus_hastighet, procent_over):
    
    # Skapa listor som platshållare för aktuella information
    klockslag = []
    gällande_hastighet = []
    hastighet = []
    
    # Extrahera klockslag, gällande hastighet & faktiskt hastighet
    for rad in plus_hastighet:
        klockslag.append(rad[0])
        gällande_hastighet.append(rad[2])
        hastighet.append(rad[3])
        
    # Omvandla datatyper str till int
    for i in range(len(gällande_hastighet)):
        gällande_hastighet[i] = int(gällande_hastighet[i])
        hastighet[i] = int(hastighet[i])
        
    # Plotta diagrammet med matplotlib
    plt.barh(klockslag, hastighet, height=0.4, color='green')
    plt.barh(klockslag, gällande_hastighet,height=0.8, color='black')
    plt.title("Tidpunkter där gällande hastighet överskridits med mer än " + procent_over + "%")
    plt.xlabel("Hastighet")
    plt.ylabel("Tid")
    plt.grid()
    plt.show()


### deluppgift 4c ###
'''
Specificerar vägnummer och kommun 
för överträdelserna med hjälp av
listorna som argument
'''
def hastighetkommun(plus_hastighet, plats_data):
    
    lista = plus_hastighet
    
    # Swappa värden för korrekt utskrift enligt kolumner
    for i in lista:
        i[4] = i[3]
        
    # Swappa värden för korrekt utskrift enligt kolumner
    for j in lista:
        for k in plats_data:
            if j[1] == k[0]:
                j[2] = k[3]
                j[3] = k[2]
    
    # Skriv ut output enligt uppgiftsbeskrivningen
    print("Hastighetsöverträdelserna skedde vid följande " + str(len(plushastighet)) + " platser.")
    print(f'Tid  \t  MätplatsID  \t  Kommun  \t  Vägnummer  \t  Hastighet')
    print("-------------------------------------------------------------------")    
    
    for rad in lista:
        print(f'{rad[0]:10} {rad[1]:14} {rad[2]:10} {rad[3]:>11} {rad[4]:>15}')


# Huvudprogram som anropar funktionerna i uppgift a, b & c
plushastighet, procentover = hastighetplus(kameradata)
hastighetplusdiagram(plushastighet, procentover)
hastighetkommun(plushastighet, platsdata)
```

```python
import numpy as np

'''
Rita diagram som visar summan av antal fordon 
som kamerorna har registrerat över alla mätplatser 
i alla kommuner uppdelat per timma mellan kl. 07 och 17.
'''
def bilar_timme(kamera_data):
    
    # Deklarera en dictionary där key = tid och value = antal registrerade bilar
    dict = {}

    '''
    Hitta varje klockslag i listan "kameradata".
    Om tid (avrundad) redan finns i dictionary, addera value med +1.
    Om tiden ej redan finns, lägg till key = tid och value = 1.
    '''
    for rad in kamera_data:
        string = rad[4]
        if string[2:3] == ":":
            if string[0:2] in dict:
                tid = string[0:2]
                dict[tid] += 1
            elif string[0:2] not in dict:
                dict[string[0:2]] = 1
        else:
            continue
            
    # X och Y koordinater som plottas
    x = ['07:00','08:00','09:00','10:00','11:00','12:00','13:00','14:00','15:00','16:00','17:00']
    y = [dict['07'], dict['08'], dict['09'], dict['10'], dict['11'], dict['12'],
         dict['13'], dict['14'], dict['15'], dict['16'], dict['17']]
   
    # Plotta tabellen med hjälp av matplotlib
    plt.plot(x, y)
    plt.yticks(np.arange(0, 200, 25))
    plt.title('Antal fordon per timme mellan kl. 07:00 och 17:00')
    plt.xlabel('Tid')
    plt.ylabel('Antal registrerade fordon')
    plt.grid()
    plt.show()
    
# Anropar funktionen bilar_timme(kamera_data)
bilar_timme(kameradata)
```
